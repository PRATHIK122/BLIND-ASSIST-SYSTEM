```python
from google.colab import files
```

```python
upload=files.upload()
```

```python
from transformers import pipeline
captioner=pipeline("image-text-to-text",model="nlpconnect/vit-gpt2-image-captioning")
image_path=list(uploaded.keys())[0]
result=captioner(image_path)
text=result[0]['generated_text']
print("Detected:",text)
```

```python
from transformers import VisionEncoderDecoderModel,ViTImageProcessor,AutoTokenizer
from PIL import Image
from gtts import gTTS
import torch
model=VisionEncoderDecoderModel.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
feature_extractor=ViTImageProcessor.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
tokenizer=AutoTokenizer.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
image_path=list(upload.keys())[0]
image=Image.open(image_path).convert("RGB")
pixel_value=feature_extractor(images=image,return_tensors="pt").pixel_values
output_ids=model.generate(pixel_value,max_length=16,num_beams=4)
text=tokenizer.decode(output_ids[0],skip_special_tokens=True)
print("Detected",text)
tts=gTTS(text=text,lang="en")
tts.save("output.mp3")
print("Audio saved as output.mp3")
```
